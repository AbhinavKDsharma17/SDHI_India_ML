Please note that data availability varies across years, and some years do not have reported values.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Example data (replace with your actual data)
years = np.array([2006, 2011, 2015])  # Known years
ger_values = np.array([23.5, 24.3, 26.3])  # Corresponding GER values

# Create a linear regression model
model = LinearRegression()
model.fit(years.reshape(-1, 1), ger_values)

# Predict GER for missing years
missing_years = np.array([2007, 2008, 2009, 2010, 2012, 2013, 2014])
predicted_ger = model.predict(missing_years.reshape(-1, 1))

# Plot the data and best-fit line
plt.scatter(years, ger_values, color='blue', label='Known Data')
plt.plot(years, model.predict(years.reshape(-1, 1)), color='red', label='Best-Fit Line')
plt.scatter(missing_years, predicted_ger, color='green', label='Predicted Data')
plt.xlabel('Year')
plt.ylabel('Tertiary GER (%)')
plt.title('Tertiary GER in India')
plt.legend()
plt.show()

# Print predicted GER for missing years
for year, ger in zip(missing_years, predicted_ger):
    print(f"Predicted GER for {year}: {ger:.2f}%")
```

## Health Index(ùë•2)

the life expectancy in India from 2006 to 2015. Life expectancy at birth reflects the average number of years a newborn is expected to live if current mortality rates remain constant throughout their lifetime.

Here are the historical life expectancy values for India during that period:

2006: 64.72 years 

2007: Not available.

2008: Not available.

2009: Not available.

2010: Not available.

2011: 66.43 years 

2012: Not available.

2013: Not available.

2014: Not available.

2015: 68.37 years 

India has made significant progress in improving life expectancy over the years, adding approximately 20 years from 1970-75 to 2015-19 23. This positive trend reflects advancements in healthcare, nutrition, and overall well-being.

```python
x = [2006, 2011, 2015]
y = [64.72, 66.43, 68.37]

n = 3  # number of data points

# Calculate sums
sum_x = sum(x)
sum_y = sum(y)
sum_xy = sum(xi * yi for xi, yi in zip(x, y))
sum_x_squared = sum(xi**2 for xi in x)

# Calculate slope (m) and y-intercept (b)
m = (n * sum_xy - sum_x * sum_y) / (n * sum_x_squared - sum_x**2)
b = (sum_y - m * sum_x) / n

# Predict values for the missing years
years_to_predict = [2007, 2008, 2009, 2010, 2012, 2013, 2014]
predicted_values = [m * year + b for year in years_to_predict]

# Display the predicted values
for year, value in zip(years_to_predict, predicted_values):
    print(f"{year}: {value:.2f} years")
```

## Employment Rate(ùë•3)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Known data (replace with actual values)
years_known = np.array([2006, 2011, 2015])
unemployment_rate_known = np.array([8.63, 8.17, 7.92])

# Create polynomial features (degree 2 for quadratic regression)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(years_known.reshape(-1, 1))

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, unemployment_rate_known)

# Predict unemployment rate for missing years
years_missing = np.array([2007, 2008, 2009, 2010, 2012, 2013, 2014])
X_missing = poly.transform(years_missing.reshape(-1, 1))
predicted_unemployment_rate = model.predict(X_missing)

# Print predicted unemployment rate for missing years
for year, rate in zip(years_missing, predicted_unemployment_rate):
    print(f"Predicted Unemployment Rate for {year}: {rate:.2f}%")

# Plot the regression curve
plt.scatter(years_known, unemployment_rate_known, color='blue', label='Known Data')
plt.plot(years_known, model.predict(X_poly), color='red', label='Polynomial Regression')
plt.scatter(years_missing, predicted_unemployment_rate, color='green', label='Predicted Data')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate (%)')
plt.title('Unemployment Rate vs. Year (Polynomial Regression)')
plt.legend()
plt.show()
```

```python
# Given unemployment rate data
unemployment_rate_data = [8.63, 8.52, 8.43, 8.33, 8.25, 8.17, 8.10, 8.03, 7.97, 7.92]

# Calculate employment rate for each year
employment_rate_data = [1 - (unemployment_rate / 25) for unemployment_rate in unemployment_rate_data]

print('Employment Rate(ùë•3)')

# Display the results
for year, employment_rate in zip(range(2006, 2016), employment_rate_data):
    print(f"{year}: {employment_rate:.4f}")
```

## GNNP(ùë•4)

Purchasing Power Parity (PPP) per capita of India from 2006 to 2015. PPP per capita represents the adjusted value of goods and services produced within an economy, considering the relative cost of living and inflation rates.
Here are the historical PPP per capita values for India during that period:

2006: $ 780 (international dollars) 

2007-2014: Not available.

2015: $1,590 (international dollars) 

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Given data
years_available = np.array([2006, 2015])
values_available = np.array([780, 1590])

# Reshape data for sklearn
years_available = years_available.reshape(-1, 1)

# Create a linear regression model
regression_model = LinearRegression().fit(years_available, values_available)

# Years for prediction (2007-2014)
years_to_predict = np.array([2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014]).reshape(-1, 1)

# Predict values for the missing years
predicted_values = regression_model.predict(years_to_predict)

print('GNNP(ùë•4)')
 
print(f"{2006}: ${780}")

# Display the predicted values
for year, value in zip(range(2007, 2016), np.concatenate([predicted_values, [1590]])):
    print(f"{year}: ${value:.2f}")

print(f"{2015}: ${1590}")    
```
